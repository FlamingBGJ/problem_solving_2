### Step 4. $10.00 이상인 제품이 몇 개 있습니까?
#### 같은 항목이면 비싼 물품 기준

```python
#float("$10.05"[1:])
"$10.05".split("$")[1]
```

```python
def sp_d(x):
    tmp = x.split("$")[1]
    return float(tmp)
```

```python
chipo_4 = chipo[(chipo.item_price.apply(sp_d)>10) & (chipo.quantity == 1) ]
chipo_4.groupby(by="item_name").min()
```

```python
len(chipo_4.item_name.unique())
```

```python
# 잘못된 풀이
chipo["only_price"] = chipo["item_price"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
chipo["int_price"] = pd.to_numeric(chipo["only_price"])
chipo["price"] = chipo["int_price"]/100
chipo.drop(["only_price","int_price","item_price"],axis = 1, inplace=True)
mask = chipo["price"]>10.00
chipo[mask]
```
